```python
import os
import re
```

```python
stopwords = open('../Data/stopwords.txt', 'r', encoding='utf8')
stopwordslist = stopwords.readlines()
stopwords.close()
```

```python
print(stopwordslist)
```

```python
sin_stopwords = []
for word in range(0, len(stopwordslist)):
    temp = re.sub('[\n]', ' ', stopwordslist[word])
    temp = re.sub('r[a-zA-A0-9]', ' ', temp)
    temp = temp.strip()
    sin_stopwords.append(temp)
```

```python
stemmer = open('../Data/stemmer.txt', 'r', encoding="utf8")
stemmerlist = stemmer.readlines()
stemmer.close()
```

```python
print(len(stemmerlist))
print(stemmerlist[1].split())
```

```python
word0 = []
word1 = []

for i in range(0, len(stemmerlist)):
    x = stemmerlist[i].split()[0]
    y = stemmerlist[i].split()[1]
    word0.append(x)
    word1.append(y)
    print(f"status = {i} / {len(stemmerlist)}")
```
